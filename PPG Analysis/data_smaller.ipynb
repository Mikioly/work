```python
import pandas as pd

# 读取原始数据
data = pd.read_csv('test_peek_data.csv', delimiter=',')
data
```

```python
# 定义函数以将每个组的数据量缩减到原来的 3/4
def reduce_group_size(group):
    return group.sample(frac=0.25)  # 保留 75% 的数据

# 根据 label 分组并应用函数
reduced_data = data.groupby('label', group_keys=False).apply(reduce_group_size)
reduced_data
```

```python
# 保存处理后的数据到新的 CSV 文件
reduced_data.to_csv('new_test_peek_data.csv', index=False)
```
